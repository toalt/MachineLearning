### 1. Implementierung der Hauptkomponentenanalyse <br>  Teil A

```python
#imports
import pandas as pd
import numpy.linalg
import numpy as np
import pylab as pl
from pandas.tools.plotting import scatter_matrix
```

```python
#Boston Housing data set
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
#Spaltendefinieren
cols =['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','TGT']
#Download
bHousing = pd.read_csv(url, sep=' ',skipinitialspace=True,header =None,names =cols,index_col = False)
origin = bHousing.copy()
```

```python
bHousing.head()
```

```python
#Zentrierung -> Schwerpunkt liegt am Ursprung

#Mittelwert -> Normalisierung
bHousingMean = bHousing.mean()

#Zentrierung auf alle Datensätze anwenden
for i in cols:
    bHousing[i] = bHousing[i] - bHousingMean[i]
```

```python
bHousing.head()
```

```python
#Durch Transposition des Arrays wird die Designmatrix erstellt.
bHousing = bHousing.transpose()
```

```python
#Lösung des Eingenwertsproblem -> Singulärwertzerlegung von X (Darstellung als Produkt in drei Matrizen).
#numpy.linalg.svd(a, full_matrices=1, compute_uv=1)
u,d,v = svd(bHousing,full_matrices=False,compute_uv=True)
```

```python
#Die Projektionen ai der Daten xi auf die ersten r Basisvektoren qj 
#(d.h die ersten r Hauptkomponenten) sind die die ersten r Spalten der n × d-Matrix UD.

#Hauptkomponenten = np.dot(u,d)-> wurde mit compute_uv=True gemacht
print "Anzahl ersten r Hauptkomponenten : r = " ,v.shape[0] 
print "Die ersten r Hauptkomponenten : r = ", u #Hauptkomponenten
```

### Teil B

```python
#Die Kovarianzmatrix enthält auf der Hauptdiagonalen die Varianzen der einzelnen Komponenten des Zufallsvektors. 
#Varianz ist die mittlere quadratische Abweichung !

#Eigenwerte der Kovarianzmatrix
Eigenwerte =[]
for i in range(0, len(d)):
        # Varianz quadratische mittlere Abweichung
        Eigenwerte.append((d[i]*d[i]))
```

```python
#Anteil an der Gesamtvarianz
Varianzanteil=[]
for i in range(0, len(Eigenwerte)): 
    #(sum(Eigenwerte))
    anteilProzent = (Eigenwerte[i])/sum(Eigenwerte)
    Varianzanteil.append(anteilProzent*100);
```

```python
# Kummulative erklärte Varianz, d.h. welchen Varianzanteil die ersten n Komponenten zusammen erklären.
KummuliertVarAnteil = []
for i in range(0, len(Varianzanteil)):
    KummuliertVarAnteil.append(sum(Varianzanteil[0:i+1]));
```

```python
Resulttable = pd.DataFrame()
Resulttable['Eigenwerte']=Eigenwerte
Resulttable['Varianzanteil (erklärte Varianz)']=Varianzanteil
Resulttable['Kummulativer Anteil der erklärten Varianz']=KummuliertVarAnteil
Resulttable
```

### Teil C

<b>Korrelationskoeffizienten der ersten drei Hauptkomponenten mit 
den ursprünglichen Variablen

```python
#Produkt-Moment-Korrelation (Bravais/Pearson)
#Werte zwischen -1 und +1  -> bei 0 keine lineare Abhängigkeit

correlation = []
for j in range(0,3):
    newLine=[]
    for i in cols:
        #Ursprüngliche Variable in bHousing
        newLine.append(np.corrcoef(origin[i],v[j]))
    correlation.append(newLine)



#Eine negative Korrelation bedeutet einen gegenläufigen Zusammenhang.
#                                     Steigt a, fällt b und umgekehrt.
#Eine positive Korrelation bedeutet einen gleichsinnigen Zusammenhang zwischen a und b.
#                               Steigt a, steigt b, fällt a, fällt auch b und umgekehrt.

#Nicht mit sich selbst 
```

```python
pd.DataFrame(correlation,columns=cols)
```

### Teil D

<b>Die erten beiden Hauptkomponenten als Scatterplot -> Streudiagramm.<br>
<b>Oberhalb und unterhalb des Medians aller Hauspreise farbliche Abgrenzung


```python
Hauptkomponenten = u*d # == np.dot(u,diag(d)St
```

```python
Hauptkomponenten.shape
```

```python
# Kaufpreis Median 
KaufMedian = median(origin['TGT'])
print 'Kaufpreis Median: ',KaufMedian
```

```python
# Target Werte color label
colorLabel = []
red = [1,0,0]
blue = [0,0,1]
for i in range (0, len(origin)):
    #rot oberhalb des Medians
    if (origin['TGT'][i] > KaufMedian):
        colorLabel.append(red)
    #blue unterhalb des Median
    else:
        colorLabel.append(blue)
```

```python
#Streudiagramm erstellen
scatterdataHaupt = np.zeros((len(origin),2))

for i in range(0, len(origin)):
    for j in range(0,2):
        scatterdataHaupt[i][j] = np.dot(Hauptkomponenten[:,j], origin.iloc[i,:])
```

```python
plt.scatter(scatterdataHaupt[:,0], scatterdataHaupt[:,1], c=colorLabel);
```
