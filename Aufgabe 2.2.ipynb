```python
# Hauptkomponentenanalyse
def PCA(array_np):
    Data = copy(array_np)
    for i in range(0, len(array_np)):
        mean = np.mean(array_np[i])
        for j in range(0,len(array_np[i])):
            array_np[i][j] -= mean
    
    Data = getInForm(Data)
    return np.linalg.svd(Data, full_matrices=True)
```
